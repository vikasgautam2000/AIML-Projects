4. Statistical Analysis: [10 Marks]

A. Is there any significant difference between Funds raised by companies that are still operating vs companies that closed down? [1 Mark]



```python
#There is no significant difference between funds raised by companies that ate still operating vs companies that closed down.
```

B. Write the null hypothesis and alternative hypothesis. [1 Mark]



```python
#H0: There is no significant difference between funds raised by companies that ate still operating vs companies that closed down.

#Ha: There is significant difference between funds raised by companies that ate still operating vs companies that closed down.
```

C. Test for significance and conclusion [1 Mark]



```python
comp_operational = data_company[data_company['OperatingState'] == "Operating"]
comp_closed = data_company[data_company['OperatingState'] == "Closed"]
tstat,pval = stats.ttest_ind(comp_operational['Funding'],comp_closed['Funding'])
if pval < 0.05:
 print(f'With a p-value of {round(pval,4)} the difference is significant. aka |We reject the null| Hence we conclude There is significant difference between funds raisedby companies that ate still operating vs companies that closed down.')
else:
 print(f'With a p-value of {round(pval,4)} the difference is not significant. aka |We fail to reject the null| Hence we conclude There is no significant difference between funds raised by companies that ate still operating vs companies that closed down.')
```
